```python
df['age'] = df['age'].fillna(df['age'].mean())
df['bp'] = df['bp'].fillna(df['bp'].mean())
df['sg'] = df['sg'].fillna(df['sg'].mean())
df['al'] = df['al'].fillna(df['al'].mean())
df['su'] = df['su'].fillna(df['su'].mean())
df['bgr'] = df['bgr'].fillna(df['bgr'].mean())
df['bu'] = df['bu'].fillna(df['bu'].mean())
df['sc'] = df['sc'].fillna(df['sc'].mean())
df['sod'] = df['sod'].fillna(df['sod'].mean())
df['pot'] = df['pot'].fillna(df['pot'].mean())
df['hemo'] = df['hemo'].fillna(df['hemo'].mean())
```

```python
df['pcc'] = df['pcc'].fillna(df['pcc'].mode()[0])
df['pcv'] = df['pcv'].fillna(df['pcv'].mode()[0])
df['wc'] = df['wc'].fillna(df['wc'].mode()[0])
df['rc'] = df['rc'].fillna(df['rc'].mode()[0])

```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.drop('id', axis=1, inplace=True)
```

```python
# Strip whitespace or tabs from all values in the 'classification' column
df['classification'] = df['classification'].str.strip()

# Now, check the counts again
target_counts = df['classification'].value_counts()

print(target_counts)


```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.boxplot(x='age', y='classification', data=df)
plt.xticks(rotation=45)

plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder

categorical_cols = df.select_dtypes(include='object').columns

le = LabelEncoder()

for col in categorical_cols:
    df[col] = le.fit_transform(df[col])

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns

plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols):
    plt.subplot(3, 4, i + 1)
    sns.boxplot(y=df[col])
    plt.title(f'Boxplot of {col}')
plt.tight_layout()
plt.show()
```

```python
corr = df.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Feature Correlation Heatmap')
plt.show()
```

```python
Q1 = df[numerical_cols].quantile(0.25)
Q3 = df[numerical_cols].quantile(0.75)
IQR = Q3 - Q1

df_no_outliers = df[~((df[numerical_cols] < (Q1 - 1.5 * IQR)) | (df[numerical_cols] > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
cor_matrix = df_no_outliers.corr().abs()
upper = cor_matrix.where(np.triu(np.ones(cor_matrix.shape), k=1).astype(bool))
to_drop = [column for column in upper.columns if any(upper[column] > 0.9)]

df_selected = df_no_outliers.drop(columns=to_drop)
```

```python
df=df_selected
```

```python
df.shape
```

```python
from sklearn.feature_selection import SelectKBest, f_classif

X = df.drop('classification', axis=1)
y = df['classification']

selector = SelectKBest(score_func=f_classif, k=8)
X_new = selector.fit_transform(X, y)

x = X.columns[selector.get_support()]
print("Selected Features:")
print(x)
```

```python
x = X[x] 
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
print("X_train \n",X_train.head())
print("X_test \n ",X_test.head())
print("y_train \n",y_train.head())
print("y_test \n",y_test.head())
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

model = LogisticRegression()
model.fit(X_train_scaled, y_train)

# Predict
y_pred = model.predict(X_test_scaled)

# Evaluate
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred,normalize="pred")
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=7)
knn.fit(X_train, y_train)
y_pred_knn = knn.predict(X_test)
print("KNN Accuracy:", accuracy_score(y_test, y_pred_knn))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred_knn,normalize="pred")
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
y_pred_dt = dt.predict(X_test)
print("Decision Tree Accuracy:", accuracy_score(y_test, y_pred_dt))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred_dt,normalize="pred")
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=100)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
print("Random Forest Accuracy:", accuracy_score(y_test, y_pred_rf))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred_rf,normalize="pred")
```
